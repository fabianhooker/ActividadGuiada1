```python
#Devolucion cambio

def cambio_moneda(CANTIDAD,SISTEMA):
    print("SISTEMA: ")
    print(SISTEMA)
     #CREA LISTA DE CEROS (TANTOS LA DIMESIÃ“N DE SISTEMA)
    SOLUCION=[0 for i in range(len(SISTEMA))]
    VALOR_ACUMULADO=0
    
    for i in range(len(SISTEMA)):
        monedas= int((CANTIDAD-VALOR_ACUMULADO)/SISTEMA[i])
        SOLUCION[i]=monedas
        VALOR_ACUMULADO+=monedas*SISTEMA[i]
        if VALOR_ACUMULADO==CANTIDAD: break
        
    return SOLUCION

SISTEMA=[25,10,5,1] 
cambio_moneda(54,SISTEMA)
```

```python
import math

def closest_points_1d_brute_force(points):
    n = len(points)
    if n < 2:
        return None
    
    min_distance = float('inf')
    closest_pair = None
    
    for i in range(n - 1):
        for j in range(i + 1, n):
            distance = abs(points[i] - points[j])
            if distance < min_distance:
                min_distance = distance
                closest_pair = (points[i], points[j])
    
    return closest_pair

points_1d = [3403, 4537, 9089, 9746, 7259, 9653]
closest_pair_1d = closest_points_1d_brute_force(points_1d)
print("1D: Closest pair:", closest_pair_1d)
```

```python
import math

def euclidean_distance(point1, point2):
    if len(point1) == 1:  # 1D
        return abs(point1[0] - point2[0])
    elif len(point1) == 2:  # 2D
        return math.sqrt((point1[0] - point2[0]) * 2 + (point1[1] - point2[1]) * 2)
    elif len(point1) == 3:  # 3D
        return math.sqrt((point1[0] - point2[0]) * 2 + (point1[1] - point2[1]) * 2 + (point1[2] - point2[2]) ** 2)
    else:
        raise ValueError("Unsupported dimension")

def closest_points_1d_divide_and_conquer(points):
    def divide_and_conquer(points):
        n = len(points)
        if n <= 1:
            return None
        
        if n == 2:
            return points
        
        points.sort()  # Ordenar los puntos en orden ascendente
        
        min_distance = float('inf')
        closest_pair = None
        
        for i in range(n - 1):
            distance = abs(points[i] - points[i + 1])
            if distance < min_distance:
                min_distance = distance
                closest_pair = (points[i], points[i + 1])
        
        return closest_pair
    
    return divide_and_conquer(points)

def closest_points_2d_divide_and_conquer(points):
    def divide_and_conquer(points):
        n = len(points)
        if n <= 1:
            return None
        
        if n == 2:
            return points
        
        points.sort(key=lambda p: p[0])  # Ordenar los puntos por su coordenada X
        
        mid = n // 2
        mid_point = points[mid]
        
        left_points = points[:mid]
        right_points = points[mid:]
        
        left_closest = divide_and_conquer(left_points)
        right_closest = divide_and_conquer(right_points)
        
        if left_closest is None:
            closest_pair = right_closest
        elif right_closest is None:
            closest_pair = left_closest
        else:
            distance_left = euclidean_distance(left_closest[0], left_closest[1])
            distance_right = euclidean_distance(right_closest[0], right_closest[1])
            
            if distance_left < distance_right:
                closest_pair = left_closest
            else:
                closest_pair = right_closest
        
        strip_points = [point for point in points if abs(point[0] - mid_point[0]) < euclidean_distance(closest_pair[0], closest_pair[1])]
        strip_points.sort(key=lambda p: p[1])  # Ordenar los puntos dentro de la banda por su coordenada Y
        
        for i in range(len(strip_points) - 1):
            for j in range(i + 1, min(i + 7, len(strip_points))):
                distance = euclidean_distance(strip_points[i], strip_points[j])
                if distance < euclidean_distance(closest_pair[0], closest_pair[1]):
                    closest_pair = (strip_points[i], strip_points[j])
        
        return closest_pair
    
    return divide_and_conquer(points)

def closest_points_3d_divide_and_conquer(points):
    def divide_and_conquer(points):
        n = len(points)
        if n <= 1:
            return None
        
        if n == 2:
            return points
        
        points.sort(key=lambda p: p[0])  # Ordenar los puntos por su coordenada x
        
        mid = n // 2
        mid_point = points[mid]
        
        left_points = points[:mid]
        right_points = points[mid:]
        
        left_closest = divide_and_conquer(left_points)
        right_closest = divide_and_conquer(right_points)
        
        if left_closest is None:
            closest_pair = right_closest
        elif right_closest is None:
            closest_pair = left_closest
        else:
            distance_left = euclidean_distance(left_closest[0], left_closest[1])
            distance_right = euclidean_distance(right_closest[0], right_closest[1])
            
            if distance_left < distance_right:
                closest_pair = left_closest
            else:
                closest_pair = right_closest
        
        strip_points = [point for point in points if abs(point[0] - mid_point[0]) < euclidean_distance(closest_pair[0], closest_pair[1])]
        strip_points.sort(key=lambda p: p[1])  # Ordenar los puntos dentro de la banda por su coordenada y
        
        for i in range(len(strip_points) - 1):
            for j in range(i + 1, min(i + 7, len(strip_points))):
                distance = euclidean_distance(strip_points[i], strip_points[j])
                if distance < euclidean_distance(closest_pair[0], closest_pair[1]):
                    closest_pair = (strip_points[i], strip_points[j])
        
        return closest_pair
    
    return divide_and_conquer(points)

# Ejemplo de uso
points_1d = [3403, 4537, 9089, 9746, 7259, 9653]
closest_pair_1d = closest_points_1d_divide_and_conquer(points_1d)
print("1D: Closest pair:", closest_pair_1d)


```
